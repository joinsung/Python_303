```python
t=np.arange(0,12,1)
y1=np.sin(t)
y2=np.cos(t)

plt.subplot(2,1,1)
plt.plot(t,y1,'r--',marker='*')
plt.plot(t,y2,'go-')
plt.grid()
plt.figure(figsize=(12,6))
```
